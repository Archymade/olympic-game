# Data cleaning

```python
#Checking for duplicate values in Summer data
sports = summer_df['Sport'].unique()
len(sports)
```

```python
#There are 43 Sports
sports
```

```python
Discipline = summer_df['Discipline'].unique()
len(Discipline)
```

```python
Discipline
```

```python
summer_df['Discipline'] = summer_df['Discipline'].str.lower()
Discipline = summer_df['Discipline'].unique()
Discipline.sort()
Discipline
```

```python
#Replacing duplicates in Discipline Column
summer_df = summer_df.replace('beach volley.', 'beach volleyball')
summer_df = summer_df.replace('bmx', 'cycling bmx')
summer_df = summer_df.replace('artistic g.', 'gymnastics artistic')
summer_df = summer_df.replace('rhythmic g.', 'gymnastics rhythmic')
summer_df = summer_df.replace('modern pentath.', 'modern pentathlon')
summer_df = summer_df.replace('synchronized s.', 'synchronized swimming')
summer_df = summer_df.replace('wrestling free.', 'wrestling freestyle')
summer_df = summer_df.replace('wrestling gre-r', 'wrestling greco roman')
```

```python
Discipline = summer_df['Discipline'].unique()
len(Discipline)
```

```python
#Merging Summer and Dictionary Dataframe
summer_df = pd.merge(summer_df,dictionary, left_on='Country', right_on='Code')
summer_df
```

```python
summer_df = summer_df.drop(["Country_x"], axis=1)
```

```python
summer_df.rename(inplace=True, columns={'Country_y':'Country'})
```

```python
summer_df.head()
```

```python
summer_df.info
```

```python
summer_df.isnull().sum()
```

```python
#Dropping the Population and GDP per Capita Column
summer_df = summer_df.drop(["Population", "GDP per Capita"], axis=1)
```

```python
summer_df.head()
```

```python
#Removing , in the Athlete Column
summer_df["Athlete"]=summer_df["Athlete"].str.split(', ').str[::-1].str.join(' ')
summer_df["Athlete"]=summer_df["Athlete"].str.title()
```

```python
summer_df.head()
```

# Exploratory Analysis

```python
Athlete_medal = summer_df.groupby(['Athlete'])['Medal'].count().reset_index().sort_values(by="Medal",ascending=False)
plot_data1 = Athlete_medal.head(10)
plot_data1
```

```python
#Top 10 Athletes with the Most medal
plt.figure(figsize=(10,6))
sns.barplot(x = plot_data1["Medal"], y = plot_data1['Athlete'])

plt.xlabel("Medal_count")
plt.ylabel("Athlete")
plt.title("Top 10 Athlete with most Olympic medals")
```

```python
df_gold = summer_df.loc[summer_df['Medal'] == 'Gold']
df_gold
```

```python
#Athlete with most Gold
Athlete_gold = df_gold.groupby(['Athlete'])['Medal'].count().reset_index().sort_values(by="Medal",ascending=False)
Athlete_gold = Athlete_gold.head(10)
Athlete_gold 
```

```python
Athlete_gold.rename(inplace=True, columns={'Medal':'Gold'})
```

```python
medals_athlete = summer_df.groupby(['Athlete','Medal'])["Sport"].count().reset_index().sort_values(by='Sport',ascending=False)
medals_athlete = medals_athlete.pivot('Athlete','Medal','Sport').fillna(0)
medals_athlete
```

```python
top_athletes=medals_athlete.sort_values(by="Gold",ascending=False)[:11]
top_athletes
```

```python
plt.figure()
top_athletes.plot.barh()
fig=plt.gcf()
fig.set_size_inches(10,10)
plt.title("Medals Distribution of Top Athletes for Summer Olympics")
plt.show()
```
