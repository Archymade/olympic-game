```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
code_summer_df = pd.read_csv("https://raw.githubusercontent.com/HDSC-21/olympic-game/main/sources/summer.csv", error_bad_lines=False)
code_winter_df = pd.read_csv("https://raw.githubusercontent.com/HDSC-21/olympic-game/main/sources/winter.csv", error_bad_lines=False)
code_dictionary_df = pd.read_csv("https://raw.githubusercontent.com/HDSC-21/olympic-game/main/sources/dictionary.csv", error_bad_lines=False)
```

```python
code_summer_df.head()
```

```python
code_winter_df.head()
```

```python
#Merging data sets **** checkingfor best merging method
winter_df = pd.merge(code_winter_df,code_dictionary_df, left_on='Country', right_on='Code')
winter_df
```

```python

column = {
   'Country': 'Code',
}
code_winter_df=code_winter_df.rename(columns = column)
code_summer_df=code_summer_df.rename(columns = column)
```

```python
code_dictionary_df.head()
```

```python
#merge for the winter
winter_df = pd.merge(code_winter_df,code_dictionary_df, on='Code')
winter_df
```

```python
#merge summer data sets
summer_df = pd.merge(code_summer_df,code_dictionary_df, on='Code')
summer_df
```

```python
#concatination of both data sets
olympic_df = pd.concat([summer_df,winter_df],keys=['summer','winter'])
olympic_df
```
